```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from itertools import cycle
import matplotlib.ticker as ticker
from matplotlib.ticker import FuncFormatter

# 各種名称の設定
filename = 'your_filename' #インプットファイル名、拡張子もつける
fontname = 'Times New Roman' #フォント名
title_name = 'your_title_name' #タイトル
plot_name = 'your_plot_name' #プロットラベル
x_label_name = 'your_xlabel_name' #X軸ラベル
y_label_name = 'your_ylabel_name' #Y軸ラベル
figname = 'your_figure_name' #出力画像ファイル名、拡張子はつけない

# 各種サイズの設定
titlesize = 20
labelsize = 20 #X軸、Y軸ラベル
linesize = 2 #プロットの線の太さ
pointsize = 1.7 #散布図の点の大きさ
legendsize = 20 #凡例の文字サイズ

# フラグ設定
title_display = False
label_display = True
grid_display = True
legend_display = False
x_scale_auto = True
y_scale_auto = True
logscale_x = False
logscale_y = False
scatter_style = False

# スケールの設定 （autoに設定している場合は不要）
x_llim = 0  # 上限値
x_ulim = 100  # 下限値
y_llim = 0  # 上限値
y_ulim = 100  # 下限値

# Color set
colorset = ['darkviolet', 'red', 'blue', 'green']
colors = cycle(colorset)

# Line styles
linestyles = ['-', '--', ':', '-.']
lses = cycle(linestyles)

df = pd.read_csv(filename, header=None, sep="\s+")

# グラフの設定
plt.figure(figsize=(7, 5))

# フォント設定
plt.rcParams['font.family'] = fontname

# タイトル設定
if title_display:
    plt.title(title_name, fontsize=titlesize)

# x軸とy軸のラベル設定
if label_display:
    plt.xlabel(x_label_name, fontsize=labelsize)
    plt.ylabel(y_label_name, fontsize=labelsize)

# 軸の範囲設定
if not x_scale_auto:
    plt.xlim(x_llim, x_ulim)
if not y_scale_auto:
    plt.ylim(y_llim, y_ulim)

# ログスケールの設定
if logscale_x:
    plt.xscale('log')

if logscale_y:
    plt.yscale('log')

# グリッドの表示
if grid_display:
    plt.grid(True, linestyle='--', color='gray', alpha=0.5)

# プロット
if scatter_style:
    plt.scatter(df[0], df[1], color=next(colors), s=pointsize)
else:
    plt.plot(df[0], df[1], label=plot_name, linestyle=next(lses), color=next(colors), linewidth=linesize)

# 複数のプロットを追加する場合の例
# for i in range(2, df.shape[1]):
#     plt.plot(df[0], df[i], label=f'Label {i}', linestyle=next(lses), color=next(colors), linewidth=2)

# 凡例の設定
if legend_display:
    plt.legend(loc='upper right', fontsize=legendsize, fancybox=True, framealpha=0.5)

# ファイルに保存
plt.savefig(f'{figname}.png')
plt.savefig(f'{figname}.pdf')

# プロット表示
plt.show()
```
